# male bangla tts trial

```python
!tts --model_name tts_models/bn/custom/vits-male --text "আমার সোনার বাংলা, আমি তোমায় ভালোবাসি"
```

```python
from TTS.api import TTS

tts = TTS(model_name="tts_models/bn/custom/vits-male")
tts.tts_to_file(text="আমার সোনার বাংলা, আমি তোমায় ভালোবাসি", file_path="output_bn.wav")

# TTS with voice conversion to a reference speaker in `target_speaker_male.wav`
#https://github.com/mobassir94/comprehensive-bangla-tts/blob/main/target_speaker_male.wav
tts.tts_with_vc_to_file(text="আমার সোনার বাংলা, আমি তোমায় ভালোবাসি", speaker_wav="/content/target_speaker_male.wav", file_path="output_cloned.wav")
```
