```python

def iitm_to_ljspeech(file):
    annotations = file.read()
    file.close()
    annotations = annotations.split("\n")
    print(len(annotations))
    texts = []
    for idx in range(len(annotations)-1):
        # using partition()
        # String till Substring
        my_list1 = annotations[idx].partition('"')[0]
        my_list1 = my_list1.replace('(','') 
        my_list = re.findall(r'"([^"]*)"', annotations[idx])
        annotations[idx] = my_list1.strip() + '|' + my_list[0].strip()
        texts.append(my_list[0].strip())
    return annotations,texts
```

```python
read_female = open("../input/bn-indictts/IndicTTS_Phase2_Bengali_fem_Speaker1_mono/mono/txt.done.data", "r")

female_annotations,male_texts = iitm_to_ljspeech(read_female)
female_annotations[0:5]
```

# Male Data Processing

```python
all_chars_male = list(set("".join(male_texts)))
all_chars_male
```

```python

punct_with_space = [i for i in all_chars_male if i in list(string.punctuation)] + [" "]
punct_with_space
```

```python
chars = [i for i in all_chars_male if i not in punct_with_space if i.strip()]
chars = "".join(chars)
chars
```

```python
punct_with_space = "".join(punct_with_space)
punct_with_space
```

```python
with open("metadata_male.txt", "w") as outfile:
    outfile.write("\n".join(male_annotations))
```
