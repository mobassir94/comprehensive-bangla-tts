```python
folder = 'upodesh_wavs'
folder1 = 'ihadis_audiobook/upodesh_wavs'
folder_path = os.path.join(os.getcwd() , folder1) 
print(folder_path)
#https://stackoverflow.com/questions/4813061/non-alphanumeric-list-order-from-os-listdir
def sorted_alphanumeric(data):
    convert = lambda text: int(text) if text.isdigit() else text.lower()
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(data, key=alphanum_key)

# files = sorted_alphanumeric(os.listdir(folder_path))
files = sorted(os.listdir(folder_path))
print(len(files))
files[0:5]
```

```python
%%time
# a big thanks to https://superuser.com/questions/1041816/combine-one-image-one-audio-file-to-make-one-video-using-ffmpeg
path = f"ihadis_videobook/{folder}"
isExist = os.path.exists(path)
if not isExist:
    os.makedirs(path)

# str = '''রাসূলুল্লাহ_saw_ও_তাঁর_সাহাবীগণের_মর্যাদা.wav''' 
# str = 'rasul saw and his sahabi.wav'
for i in range(len(files)):

#     if(files[i] != str):
#         continue
#     print(files[i])
    fpath = folder_path+'/'+files[i] 
    print(fpath)
    head, output = os.path.split(files[i])
    
    fname = output.split(".")[0]
    fname = fname.replace('\'','')
    
    fpath_new = re.sub(r'[\\*?:()&"<>|]',"_",fpath)
    os.rename(fpath, fpath_new)
    fpath_new = fpath_new.replace('\'','')
    print("working on ",fpath_new)
    fname = re.sub(r'[\\*?:()&"<>|]',"_",fname)
    out_fname = f'./{path}/ঊপদেশ_{fname}.mp4'
    cmd = f'ffmpeg -loop 1 -i ./logoCombined2.png -i {fpath_new} -c:v libx264 -tune stillimage -c:a aac -b:a 192k -pix_fmt yuv420p -shortest {out_fname}'
    os.system(cmd)
```
