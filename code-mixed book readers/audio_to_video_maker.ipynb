```python
%%time
# a big thanks to https://superuser.com/questions/1041816/combine-one-image-one-audio-file-to-make-one-video-using-ffmpeg
try:
    os.mkdir('./jalalayn_videobook')
except:
    print("exist!!")
for i in range(1,115):
    fname = f'./jalalayn_reader/{i}.wav'
    out_fname = f'./jalalayn_videobook/{i}.mp4'
    cmd = f'ffmpeg -loop 1 -i ./logoCombined2.png -i {fname} -c:v libx264 -tune stillimage -c:a aac -b:a 192k -pix_fmt yuv420p -shortest {out_fname}'
    os.system(cmd)
```
